```python
import pandas as pd 
from mlxtend.frequent_patterns import apriori,association_rules
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df=pd.read_csv("book.csv")
df.head()
```

### support=0.2,confidence=0.8

```python
frequent_itemssets=apriori(df,min_support=0.2,use_colnames=True)
```

```python
frequent_itemssets
```

```python
rules=association_rules(frequent_itemssets,metric='lift',min_threshold=0.8)
```

```python
rules
```

### support=0.1,confidence=0.6

```python
frequent_itemssets=apriori(df,min_support=0.1,use_colnames=True)
frequent_itemssets
```

```python
rules=association_rules(frequent_itemssets,metric='lift',min_threshold=0.6)
rules
```

```python
rules[rules.lift>1]
```

```python
plt.scatter(rules.support,rules.confidence)
plt.xlabel("Support")
plt.ylabel("Confidence")
```
